Libraries

```python
import numpy as np 
import pandas as pd 
import statistics as st 
import matplotlib.pyplot as plt 
```

Dataset

```python
dataset=pd.read_csv(r'C:\Users\agnih\Downloads\slr12.csv')
dataset.describe()
X=dataset.iloc[:,:-1].values
y=dataset.iloc[:,-1].values
```

Finding Corelation

```python
plt.scatter(X,y,color='blue')
plt.xlabel("Capital Investment")
plt.ylabel("Profits Gain")
plt.title("Capital VS Gains")
correla=st.correlation(X,y)
print(correla) #Showing Positive Corelation by both Statistics library and Plotting them
```

```python
from sklearn.linear_model import LinearRegression
regressor= LinearRegression()
regressor.fit(X,y)
y_pred=regressor.predict(X)
```
