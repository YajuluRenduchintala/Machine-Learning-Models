Libraries

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

Dataset

```python
dataset=pd.read_csv(r'C:\Users\agnih\Desktop\ML\Position_Salaries.csv')
print(dataset)
X=dataset.iloc[:,1:-1]
y=dataset.iloc[:,-1]
```

Training Model for Linear Regression

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X,y)
```

Training Model for Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg=PolynomialFeatures(degree=4) #Here degree = 2 represents the polynomial equation as y=b0 + b1*x1 + b2*x1^2
#Degree 4 is given as it more precise results than 2 
X_poly = poly_reg.fit_transform(X)
lin_reg_2=LinearRegression()
lin_reg_2.fit(X_poly,y)
```

Visualizing Linear Regression

```python
plt.scatter(X,y,color='red')
plt.plot(X,regressor.predict(X),color='blue')
plt.title("Truth or Bluff(Linear Regression)")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.show()
```

Visualizing Polynomial Results

```python
plt.scatter(X,y,color='red')
plt.plot(X,lin_reg_2.predict(poly_reg.fit_transform(X)),color='blue')
plt.title("Truth or Bluff (Polynomial Regression)")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.show()
```

Predicting with Linear Regression

```python
regressor.predict([[5.5]])
```

Predicting with Polynomial Regression

```python
lin_reg_2.predict(poly_reg.fit_transform([[5]]))
```
